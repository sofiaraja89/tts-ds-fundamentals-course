### Exercise 1. 

Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
````

````python
# Exercise 1

import pandas as pd
import matplotlib.pyplot as plt  
company_sales = pd.read_csv("data/company_sales_data.csv")
#company_sales
#company_sales.describe()
profit = company_sales ['total_profit'].tolist()
month  = company_sales ['month_number'].tolist()
plt.plot(month, profit, label = 'Month-wise Profit data of last year')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.xticks(month)
plt.title('Company Profit Per Month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# Exercise 3
# facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
import pandas as pd
import matplotlib.pyplot as plt  

company_sales = pd.read_csv("data/company_sales_data.csv")
month  = company_sales ['month_number'].tolist()
face_cream   = company_sales ['facecream'].tolist()
face_wash   = company_sales ['facewash'].tolist()
tooth_paste = company_sales ['toothpaste'].tolist()
bathing_soap   = company_sales ['bathingsoap'].tolist()
shampoo   = company_sales ['shampoo'].tolist()
moisturizer = company_sales ['moisturizer'].tolist()

plt.plot(month, face_cream,   label = 'Face Cream', marker='o', linewidth=3)
plt.plot(month, face_wash,   label = 'Face Wash',  marker='o', linewidth=3)
plt.plot(month, tooth_paste, label = 'Tooth Paste', marker='o', linewidth=3)
plt.plot(month, bathing_soap, label = 'Bathing Soap', marker='o', linewidth=3)
plt.plot(month, shampoo, label = 'Shampoo', marker='o', linewidth=3)
plt.plot(month, moisturizer, label = 'Moisturizer', marker='o', linewidth=3)

plt.xlabel('Month Number')
plt.ylabel('Sales Units In Number')
plt.legend(loc='upper left')
plt.xticks(month)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('All Product Sales')
plt.show()
````
