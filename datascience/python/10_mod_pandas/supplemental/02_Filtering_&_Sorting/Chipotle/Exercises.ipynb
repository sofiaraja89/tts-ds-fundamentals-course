### Step 3. Assign it to a variable called chipo.

```python
url_address = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url_address, sep = '\t')
```

### Step 4. How many products cost more than $10.00?

```python
#change item price to a float
#price = [float(value[1 : -1]) for value in chipo.item_price]
#set as price
#chipo.item_price = price
#delete duplicates in following 
#delete_copies = chipo.drop_duplicates(['item_name','quantity','choice_description']) 
#set product after deleting copies
#product = delete_copies[delete_copies.quantity == 1]
#product
#check for products whose item prive is above $10
#chipo.query('item_price > 10').item_name.nunique() 

#ANOTHER WAY TO DO IT

#Change item price from string to float 
chipo ['price_float'] = pd.to_numeric(chipo.item_price.str.slice(1))
chipo['price_float'].head()
#Setted float and get sum 
(chipo['price_float'] > 10).sum()
#Create price per item to avoid duplicates of items and item prices
chipo ['price_per_item'] = chipo.price_float / chipo.quantity
#Group prices by item name and newly made price float
price_of_products = chipo.groupby('item_name').agg({'price_per_item' : 'max'})
#Total of products that cost more than 10
(price_of_products.price_per_item > 10).sum()
```
